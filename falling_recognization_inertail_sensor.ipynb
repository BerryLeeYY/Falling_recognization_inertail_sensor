# Dataset

MobiFall dataset include the signal information from the accelerometer and gyroscope sensors of a smartphone. During the movement, the smartphone was placed in a trouser pocket with the random orientation.

11 participants were involved in the task and the target movements include 9 daily activities and 4 falling movements. 

Reference:
https://bmi.hmu.gr/the-mobifall-and-mobiact-datasets-2/

# Preprocessing data

## Extract the data from .txt

```python
def personal_information(file, Fall_condition):
    
    import pandas as pd
    # read the readme file which contain the personal information
    read_file = open(r"C:/Users/a1003/falling_prediction/MobiFall_Dataset_v1.0/readme.txt","r") 
    read_list_text = []
    for line in read_file.readlines():
      stripped_line = line.strip()
      line_list = stripped_line.split()
      read_list_text.append(line_list)
    
    # give the value to sex, height, weight and age
    list_text = []
    for line in file.readlines():
      stripped_line = line.strip()
      line_list = stripped_line.split()
      list_text.append(line_list)
    
    if list_text[5][2] == '1':
        sex = read_list_text[17][-2]
        height = read_list_text[17][-6]
        weight = read_list_text[17][-4]
        age = read_list_text[17][-8]
    if list_text[5][2] == '2':
        sex = read_list_text[18][-2]
        height = read_list_text[18][-6]
        weight = read_list_text[18][-4]
        age = read_list_text[18][-8]
    if list_text[5][2] == '3':
        sex = read_list_text[19][-2]
        height = read_list_text[19][-6]
        weight = read_list_text[19][-4]
        age = read_list_text[19][-8]
    if list_text[5][2] == '4':
        sex = read_list_text[20][-2]
        height = read_list_text[20][-6]
        weight = read_list_text[20][-4]
        age = read_list_text[20][-8]
    if list_text[5][2] == '5':
        sex = read_list_text[21][-2]
        height = read_list_text[21][-6]
        weight = read_list_text[21][-4]
        age = read_list_text[21][-8]
    if list_text[5][2] == '6':
        sex = read_list_text[22][-2]
        height = read_list_text[22][-6]
        weight = read_list_text[22][-4]
        age = read_list_text[22][-8]
    if list_text[5][2] == '7':
        sex = read_list_text[23][-2]
        height = read_list_text[23][-6]
        weight = read_list_text[23][-4]
        age = read_list_text[23][-8]
    if list_text[5][2] == '8':
        sex = read_list_text[24][-2]
        height = read_list_text[24][-6]
        weight = read_list_text[24][-4]
        age = read_list_text[24][-8]
    if list_text[5][2] == '9':
        sex = read_list_text[25][-2]
        height = read_list_text[25][-6]
        weight = read_list_text[25][-4]
        age = read_list_text[25][-8]
    if list_text[5][2] == '10':
        sex = read_list_text[26][-2]
        height = read_list_text[26][-6]
        weight = read_list_text[26][-4]
        age = read_list_text[26][-8]
    if list_text[5][2] == '11':
        sex = read_list_text[27][-2]
        height = read_list_text[27][-6]
        weight = read_list_text[27][-4]
        age = read_list_text[27][-8]
        
    # movement_type
    
    movement_type = list_text[0][0] 
    
     ## movement_type
    information_type = list_text[4][3]         
 # extract the desired  data

    
    x = []
    y = []
    z = []
    list_text = list_text[16:]
    for line in list_text:
        #ID.append(line[0][:-1])
        x.append(line[1][:-1])
        y.append(line[2][:-1])
        z.append(line[3][:-1])
    

   
    ## store the data in dataframe
    data = pd.DataFrame()
    #data['ID'] = ID
    data['x'] = x
    data['y'] = y
    data['z'] = z
    data['Fall or not']= Fall_condition
    data['sex'] = sex
    data['height'] = height
    data['weight'] = weight
    data['age'] = age
    data['movement_type'] = movement_type
    data['information_type'] = information_type
    
    
    return data
        
```

### Peak Average

```python
from scipy.signal import find_peaks


################################  P_peak of each axis for falling data ###################################

P_peak_fall_x = []
P_peak_fall_y = []
P_peak_fall_z = []

for i in range(len(total_data_fall)):
    peaks, _ = find_peaks((pd.to_numeric(total_data_fall[i]['x'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak = (pd.to_numeric(total_data_fall[i]['x'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_fall_x.append(p_mean_peak)
    
    peaks, _ = find_peaks((pd.to_numeric(total_data_fall[i]['y'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak = (pd.to_numeric(total_data_fall[i]['y'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_fall_y.append(p_mean_peak)
    
    peaks, _ = find_peaks((pd.to_numeric(total_data_fall[i]['z'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak =(pd.to_numeric(total_data_fall[i]['z'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_fall_z.append(p_mean_peak)
    
################################  P_peak of each axis for not falling data ###################################    
    
P_peak_ADL_x = []
P_peak_ADL_y = []
P_peak_ADL_z = []


for i in range(len(total_data_ADL)):
    peaks, _ = find_peaks((pd.to_numeric(total_data_ADL[i]['x'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak = (pd.to_numeric(total_data_ADL[i]['x'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_ADL_x.append(p_mean_peak)
    
    peaks, _ = find_peaks((pd.to_numeric(total_data_ADL[i]['y'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak = (pd.to_numeric(total_data_ADL[i]['y'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_ADL_y.append(p_mean_peak)
    
    peaks, _ = find_peaks((pd.to_numeric(total_data_ADL[i]['z'], errors='coerce').fillna(0)), height=0, prominence = 1)
    p_mean_peak =(pd.to_numeric(total_data_ADL[i]['z'], errors='coerce').fillna(0))[peaks].mean()
    P_peak_ADL_z.append(p_mean_peak)
    
################################  N_peak of each axis for falling data ###################################  
    
n_peak_fall_x = []
n_peak_fall_y = []
n_peak_fall_z = []

for i in range(len(total_data_fall)):
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_fall[i]['x'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_fall[i]['x'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_fall_x.append(n_mean_peak)
    
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_fall[i]['y'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_fall[i]['y'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_fall_y.append(n_mean_peak)
    
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_fall[i]['z'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_fall[i]['z'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_fall_z.append(n_mean_peak)
    
################################  N_peak of each axis for not falling data ###################################      
    
n_peak_ADL_x = []
n_peak_ADL_y = []
n_peak_ADL_z = []

for i in range(len(total_data_ADL)):
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_ADL[i]['x'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_ADL[i]['x'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_ADL_x.append(n_mean_peak)
    
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_ADL[i]['y'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_ADL[i]['y'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_ADL_y.append(n_mean_peak)
    
    peaks, _ = find_peaks(0-(pd.to_numeric(total_data_ADL[i]['z'], errors='coerce').fillna(0)), height=0, prominence = 1)
    n_mean_peak = (pd.to_numeric(total_data_ADL[i]['z'], errors='coerce').fillna(0))[peaks].mean()
    n_peak_ADL_z.append(n_mean_peak)
```
